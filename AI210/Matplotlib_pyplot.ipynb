```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame = True)
df = iris.frame

# exploratory data analysis
print(df.head())
print(df.info())
print(df.describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8, 6))
plt.scatter(x = df['sepal length (cm)'], y = df['sepal width (cm)'], color = 'red', marker = '*', alpha = 0.7)
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

```python
import seaborn as sns

df = sns.load_dataset("flights")

# exploratory data analysis
df.head()
df.info()
df.describe()
```

```python
plt.figure(figsize = (8, 6))
plt.plot(df['passengers'], marker = "o", linestyle = "dashed",
         markerfacecolor = "red", markeredgecolor = "red",
         color = "blue", label = 'line')
plt.title("Row Indices vs # of Passengers")
plt.xlabel("Row Indices")
plt.ylabel("# of Passengers")
plt.legend()

# The general trend is up, though it seems to follow a zigzag pattern
```

```python
df1 = pd.DataFrame({'year': df['year'], 'annual passengers': df['passengers']})
plt.plot(df1['year'], df1['annual passengers'], marker = "o", linestyle = "dashed",
         markerfacecolor = "red", markeredgecolor = "red",
         color = "blue", label = 'line')
plt.title("Year vs # of Passengers")
plt.xlabel("Year")
plt.ylabel("# of Passengers")
plt.legend()
```

```python
df2 = pd.DataFrame({'month': df['month'], 'monthly passengers': df['passengers']})
plt.plot(df2['month'], df2['monthly passengers'], marker = "o", linestyle = "dashed",
         markerfacecolor = "red", markeredgecolor = "red",
         color = "blue", label = 'line')
plt.title("Month vs # of Passengers")
plt.xlabel("Month")
plt.ylabel("# of Passengers")
plt.legend()
```

```python
import numpy as np

x = np.linspace(0, 10, 101)
y = np.sin(x)

plt.figure(figsize = (8, 4))
plt.plot(x, y, linewidth = 2, color = "blue", label = "sin x")
plt.title("Line Plot of sin(x)")
plt.xlabel("x points")
plt.ylabel("y points")
plt.legend()
```

```python
np.random.seed(53)

x = np.random.rand(100)
noise = np.random.randn(100) * 0.1
y = 2 * x + noise

plt.figure(figsize = (8, 6))
plt.scatter(x, y, color = "purple", marker = "o", s = 50)
plt.title("Scatter Plot: y = 2x + Noise")
plt.xlabel("x points")
plt.ylabel("y points")
```

```python
categories = ['Apple', 'Banana', 'Cherry', 'Date']
values = [10, 20, 15, 5]

plt.figure(figsize = (8, 4))
plt.bar(categories, values, color = "green")
plt.title("Vertical Bar Plot of Fruits")
plt.xlabel("Fruits")
plt.ylabel("Number")

plt.figure(figsize = (8, 4))
plt.barh(categories, values, color = "green")
plt.title("Horizontal Bar Plot of Fruits")
plt.xlabel("Number")
plt.ylabel("Fruits")
```
