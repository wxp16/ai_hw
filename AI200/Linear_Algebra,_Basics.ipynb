### Linear Algebra, Basics


To prove that if $\mathcal{W}$ is a linear subspace, then $\mathbf{0} \in \mathcal{W}$ we cam first say that since $\mathcal{W}$ is a linear subspace, by definition, there should be a spanning set for $\mathcal{W}$ whose span is equal to $\mathcal{W}$. However, things that do not pass through $0$ are not a span of any collection of vectors because the span of any collection of $n$-vectors should always contain $\mathbf{0}$ because setting the coefficients $c_1, ..., c_k$ to $0$ gives $0\mathbf{v}_1+...+0\mathbf{v}_l = \mathbf{v}$. As a result, since $\mathcal{W}$ is a linear subspace, the zero vector must be in $\mathcal{W}$.

For vectors $\mathbf{u}$ and $\mathbf{v}$, $\mathbf{u} \cdot \mathbf{v} = 0$ when the residual of $\mathbf{u}$ projected to $\mathbf{v}$ is $\mathbf{u}$ because the vectors are already orthogonal. When the residual of $\mathbf{u}$ projected to $\mathbf{v}$ is $\mathbf{u}$ that means the projection is $\mathbf{0}$. This only occurs when $\mathbf{u}$ and $\mathbf{v}$ are already orthogonal, so as a result $\mathbf{u} \cdot \mathbf{v} = 0$.

If $\{\vec{\mathbf{u}}_i\}^{r-1}_{i=0}$ is a basis of space $\mathcal{W}$ that means that our basis spans all of $\mathcal{W}$. We can use a linear combination of our basis to get to any point on $\mathcal{W}$, by definition. As a result, we can then say that any vector $\vec{v}$, as long as it is in $\mathcal{W}$ can be written as a linear combination of the basis vectors because they span that space.
